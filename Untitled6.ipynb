```python
import pandas as pd
df = pd.read_csv('tweets.csv.zip')
df.columns = df.columns.str.replace('#', '', regex=False)
df = df.applymap(lambda x: x.replace('#', '') if isinstance(x, str) else x)
df.to_csv('cleaned_data_no_hash.csv.zip', index=False)
print("CSV file cleaned, and saved to 'cleaned_data_no_hash.csv.zip'")
```
